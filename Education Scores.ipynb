```python
abbreviations = []
for name in education['Name']:
    if name in us_state_to_abbrev:
        abbreviations.append(us_state_to_abbrev[name])
    else:
        abbreviations.append(None)
abbreviations
```

```python
education['State'] = abbreviations
cleaned = education.dropna()[['Education Percentage', 'State']]
cleaned.head()
```

```python
floats = []
for i in cleaned['Education Percentage']:
    percentage = i[:5]
    floats.append(float(percentage))
floats
```

```python
cleaned['Education Percentage'] = floats
cleaned
```

```python
stars = pd.read_csv('player star ratings cleaned.csv')
stars
```

```python
stars = stars[['Score', 'ST']]
```

```python
stars = stars.dropna()
```

```python
stars = stars[stars['Score'] != 'NR']
stars
```

```python
stars['Score'] = stars['Score'].astype(float)
average_scores = stars.groupby('ST').mean()
average_scores
```

```python
average_scores['State'] = average_scores.index
average_scores
```

```python
total_scores = {}
for state in cleaned['State']:
    if state == 'PR':
        continue
    education_score = float(cleaned.loc[education['State'] == state]['Education Percentage'])
    recruiting_score = average_scores.loc[state]['Score']
    total_scores[state] = education_score * recruiting_score / 1000

total_scores
```

```python
education_and_recruiting = pd.DataFrame()
education_and_recruiting['State'] = total_scores.keys()
education_and_recruiting['Total Score'] = total_scores.values()
education_and_recruiting
```
