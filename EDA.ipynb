```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
salaries = pd.read_csv("football_salaries.csv")
```

```python
salaries
```

```python
avg_by_position = salaries[["position", "avg_year"]].groupby(["position"]).mean()
```

```python
avg_by_position.plot.bar()
```

```python
salaries.sort_values(by="avg_year", ascending=False)
```

```python
stars = pd.read_csv("player star ratings.csv")
```

```python
stars.columns
```

```python
full_names = stars['Unnamed: 6'] + " " + stars["Unnamed: 7"]
full_names
```

```python
stars["Full Name"] = full_names
```

```python
stars.columns
```

```python
merged = salaries.merge(right=stars, left_on="player", right_on="Full Name")
merged
```

```python
stars
```

```python
stars_and_salaries = merged[['position', 'player', 'total_value', 'avg_year', 'total_guaranteed', 'fully_guaranteed', 'Unnamed: 4', 'Unnamed: 5']]
stars_and_salaries
```

```python
stars_and_salaries['Star'] = stars_and_salaries['Unnamed: 4']
stars_and_salaries['Score'] = stars_and_salaries['Unnamed: 5']
stars_and_salaries = stars_and_salaries[['position', 'player', 'total_value', 'avg_year', 'total_guaranteed', 'fully_guaranteed', 'Star', 'Score']]
stars_and_salaries
```

```python
wr_salaries = stars_and_salaries.loc[stars_and_salaries['position'] == 'wide-receiver'].dropna()
wr_salaries
```

```python
wr_salaries['Score'] = wr_salaries['Score'].astype(float)
wr_salaries
```

```python
grouped = wr_salaries.groupby('player').mean()
```

```python
plt.scatter(grouped['Score'], grouped['avg_year'])
```

```python
grouped
```

```python
np.corrcoef(grouped['Score'], grouped['avg_year'])
```

```python
from scipy.stats.stats import pearsonr
```

```python
pearsonr(grouped['Score'], grouped['avg_year'])
```
